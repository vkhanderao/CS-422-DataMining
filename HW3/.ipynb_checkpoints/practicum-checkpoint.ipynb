## Problem 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data=pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx")
data.head()
```

```python
data['Country'].value_counts().sort_values()
```

```python
data.shape
```

```python
dataFrance=data[data['Country']=='France']
dataFrance=dataFrance[['InvoiceNo','Description']]
dataFrance.head()
```

```python
newData=pd.DataFrame(columns=dataFrance["Description"].unique())
newData.insert(loc=0,column="InvoiceNo",value=dataFrance["InvoiceNo"].unique())
newData.head()
```

```python
newData = newData.replace(np.nan, False)
newData.head()
```

```python
newData.shape
```

```python
for idx, row in dataFrance.iterrows():
    invoice, item= row["InvoiceNo"], row["Description"]
    newData.loc[newData['InvoiceNo'] == invoice, item]=True
```

```python
newData
```

```python
newData = newData.drop("InvoiceNo", axis=1)
```

```python
frequentItemsets = apriori(newData, min_support=0.05, use_colnames=True)
frequentItemsets.head()
```

```python
confidence = association_rules(frequentItemsets, metric="confidence", min_threshold=1)
confidence.head()
```

```python
lift = association_rules(frequentItemsets, metric="lift", min_threshold=11)
lift.head()
```

From the above tables we can see tThe rule with the highest confidence and lift are not same.<br>
***confidence*** determines how frequently items in Y appear in transactions that contain X.<br>
***Interest factor*** is able to detect meaningful patterns in the data which makes it one of the best measures to analyze indeoendence of varaibles.
